# Churn Classification in TensorFlow

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('Churn.csv')
```

```python
X = pd.get_dummies(df.drop(['Churn', 'Customer ID'], axis=1))
y = df['Churn'].apply(lambda x: 1 if x=='Yes' else 0)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)
```

```python
X_train.head()
```

```python
y_train.head()
```

### Neural Network

```python
model = Sequential()

# 2 fully-connected hidden layers
model.add(Dense(units=32, activation='relu', input_dim=len(X_train.columns)))
model.add(Dense(units=64, activation='relu'))

# because we have a binary classifier we can pass our hidden layer outputs into a sigmoid classifier
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics='accuracy')
```

### Fit, Predict, and Evaluate NN

```python
X_train = X_train.astype('float32')
y_train = y_train.astype('float32')
```

```python
# train/fit on training data
model.fit(X_train, y_train, epochs=500, batch_size=32)
```

```python
X_test = X_test.astype('float32')
# test on unseen test data
y_hat = model.predict(X_test)

# convert continuous value into binary classification. 
y_hat = [0 if val < 0.5 else 1 for val in y_hat]
```

```python
y_hat
```

```python
# we can use built-in accuracy_score function from scikit learn
accuracy_score(y_test, y_hat)
```
